```python
%load_ext autoreload
%autoreload 2

from hub.srgan import SRGAN

path = "../srgan.anec.so"
model = SRGAN(path=path)
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("data/srgan-input.jpg")[:,:,::-1]
print(img.shape)
plt.imshow(img); plt.show()
```

```python
inarr = model.preprocess(img)
inputs = [inarr]
```

```python
outputs = model.predict(inputs)
```

```python
pred = model.postprocess(outputs)
print(pred.shape)
```

```python
plt.imshow(pred); plt.show()
```
